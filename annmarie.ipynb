Now look at the relationship between affordability and the time it takes for crimes to be cleared after they're reported:

```python
df['Clearance_Time'] = pd.to_datetime(df['Clearance_Date'], format='%d-%b-%y') - pd.to_datetime(df['Report_Date'], format='%d-%b-%y')
df['Clearance_Time'] = df['Clearance_Time'].dt.days
df = df.dropna(subset=['Clearance_Time'])

clearedByArrest = df[df['Clearance_Status'] == 'C']
clearedByException = df[df['Clearance_Status'] == 'O']
notCleared = df[df['Clearance_Status'] == 'N']

# sns.distplot(df['Clearance_Time'], label="all", hist=False)
sns.distplot(clearedByArrest['Clearance_Time'], label='arrest', hist=False)
sns.distplot(clearedByException['Clearance_Time'], label='exception', hist=False)
sns.distplot(notCleared['Clearance_Time'], label='not', hist=False)
plt.xlabel('Clearance Time (Days)')
plt.ylabel('Density')
plt.title('Distribution of Clearance Time in Days')
plt.legend()
plt.show()
```

Now show what the clearance time looks like for areas with different affordability:

```python
majorityArtistRental = df[df['Rentalunitsaffordabletoaverageartist'] > 50]

minorityTechRental = df[df['Rentalunitsaffordabletoaveragetechworker'] < 50]
minorityTechOwner = df[df['Ownerunitsaffordabletoaveragetechworker'] < 50]

sns.distplot(minorityTechOwner['Clearance_Time'], label='owner', hist=False)
sns.distplot(minorityTechRental['Clearance_Time'], label='rental', hist=False)
plt.xlabel('Clearance Time (Days)')
plt.ylabel('Density')
plt.title('Distribution of Clearance Time in Days (for areas with <50% housing affordable to tech workers)')
plt.legend()
plt.show()

minorityTechRental.head()
display(len(minorityTechOwner))
display(len(minorityTechRental))
```

Something interesting to note here: there are no areas in this crime db where less than 50% of rental units are affordable to the average tech worker. Tech workers can basically afford to rent anywhere in this db. There _are_, however, 10050 crimes reported in areas where less than 50% of owner units are affordable to the average tech worker.
